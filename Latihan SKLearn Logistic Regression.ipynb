```python
import pandas as pd

# reading dataset and convert to dataframe
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.head()
```

```python
# drop unused columns
data = df.drop(columns=['User ID'])

# run one-hot encoding process using pd.get_dummies()
data = pd.get_dummies(data, dtype='int')
data
```

```python
# separate attribute with label
predictions = ['Age' , 'EstimatedSalary' , 'Gender_Female' , 'Gender_Male']
X = data[predictions]
y = data['Purchased']
```

```python
# normalization the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns=X.columns)
scaled_data.head()
```

```python
from sklearn.model_selection import train_test_split

# divide the data to train and test for each attribute and label
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)
```

```python
from sklearn import linear_model

# train model with fit function
model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
```

```python
# testing model accuracy
model.score(X_test, y_test)
```
